# Matplotlib example of workflow

```python
# Step 0: import matplotlib and get it ready to plot in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# Step 1: prepare the data
x = [1, 3, 4, 6]
y = [3, 4, 5, 7]

#Step 2: set up plot
fig, ax = plt.subplots(figsize=(10,10))

# Step 3: Plot the data
ax.plot(x, y);

# Step 4: Customize the plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# Step 5: Save and show
fig.savefig("matplotlib-results/sample-plot.png")
```

# Making figure with numpy arrays

We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# make scater plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make plot from dictionary
py_dict = {"1st": 10,
           "2nd": 20,
           "3rd": 50}
fig, ax = plt.subplots()
# ax.plot(py_dict.keys(), py_dict.values())
ax.bar(py_dict.keys(), py_dict.values())
ax.set(title = "Plot Python Dictionary",
       ylabel = "number")
```

```python
fig, ax = plt.subplots()
ax.barh(list(py_dict.keys()), list(py_dict.values()));
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# The two options for subplot

```python
# Subplot option 1
# (ax1, ax2) -> 1st row, (ax3, ax4) --> 2nd row
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))

# Plot in different axises
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(py_dict.keys(), py_dict.values())
ax4.hist(np.random.randn(1000))
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                     ncols = 2,
                     figsize = (10, 5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(py_dict.keys(), py_dict.values())
ax[1, 1].hist(np.random.randn(1000))
```
